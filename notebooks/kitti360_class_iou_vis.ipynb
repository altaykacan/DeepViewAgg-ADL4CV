```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

Copy the data from the Weights and Biases Log.

```python
val_iou = {'road': 0.6923171718512252, 'sidewalk': 0.05056567987025644, 'building': 0.3355811618471233, 'wall': 0.00010058092396470187, 'fence': 9.476156125312943e-06, 'pole': 0.13605531656041905, 'traffic light': 1e-08, 'traffic sign': 1e-08, 'vegetation': 0.6688969428045471, 'terrain': 0.36554244742442965, 'person': 1e-08, 'car': 0.24427188427911942, 'truck': 1e-08, 'motorcycle': 1e-08, 'bicycle': 1e-08}
labels = list(val_iou.keys())
scores = list(val_iou.values())
```

Create the plot for the class-wise validation mIoU scores.

```python
fig = plt.figure(figsize=(16, 8))
plt.style.use('_mpl-gallery')

bars = np.array(labels)
height = np.array(scores)* 100

x_pos = np.arange(len(labels))



plt.bar(x_pos, height, width=1, edgecolor="white", linewidth=0.7)
plt.xticks(x_pos, bars, rotation=90) # rotates the bar name
plt.ylim([0.0, 100.0])
plt.ylabel('Validation IoU')
plt.savefig("figs/classwise_val_iou_kitti.png",bbox_inches='tight')
plt.show()

```
