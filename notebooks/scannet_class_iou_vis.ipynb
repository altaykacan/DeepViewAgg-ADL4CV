```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
CLASS_LABELS = (
    "wall",
    "floor",
    "cabinet",
    "bed",
    "chair",
    "sofa",
    "table",
    "door",
    "window",
    "bookshelf",
    "picture",
    "counter",
    "desk",
    "curtain",
    "refrigerator",
    "shower curtain",
    "toilet",
    "sink",
    "bathtub",
    "otherfurniture",
)
```

Copy the data from the Weights and Biases Log.

```python
val_iou = val_miou_per_class = {0: '78.03', 1: '93.19', 2: '47.35', 3: '64.10', 4: '79.48', 5: '64.35', 6: '53.72', 7: '54.17', 8: '48.42', 9: '66.47', 10: '26.26', 11: '45.37', 12: '36.02', 13: '48.42', 14: '38.69', 15: '9.14', 16: '81.29', 17: '55.74', 18: '27.85', 19: '44.65'}
labels = [CLASS_LABELS[idx] for idx in val_iou.keys()]
scores = [ float(score) for score in val_iou.values() ]
```

Create the plot for the class-wise validation mIoU scores.

```python
fig = plt.figure(figsize=(16, 4))
plt.style.use('_mpl-gallery')

bars = np.array(labels)
height = np.array(scores)

x_pos = np.arange(len(labels))

plt.rcParams.update({'font.size': 22})

plt.bar(x_pos, height, width=1, edgecolor="white", linewidth=0.7)
plt.xticks(x_pos, bars, rotation=90) # rotates the bar name
plt.ylim([0.0, 100.0])
plt.ylabel('Validation IoU')
plt.savefig("figs/classwise_val_iou_scannet.png",bbox_inches='tight')
plt.show()

```
